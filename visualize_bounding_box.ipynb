```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def draw_bbox(image_path, bbox_path):
    # Load the image
    img = cv2.imread(image_path)
    img_height, img_width = img.shape[:2]

    # Open the YOLO bounding box file
    with open(bbox_path, 'r') as f:
        lines = f.readlines()

    # Loop over the bounding boxes
    for line in lines:
        parts = line.strip().split()

        # Parse the bounding box
        class_id = int(parts[0])
        x_center = float(parts[1]) * img_width
        y_center = float(parts[2]) * img_height
        width = float(parts[3]) * img_width
        height = float(parts[4]) * img_height

        # Calculate the top-left and bottom-right coordinates
        x_min = int(x_center - width / 2)
        y_min = int(y_center - height / 2)
        x_max = int(x_center + width / 2)
        y_max = int(y_center + height / 2)

        # Draw the bounding box onto the image
        color = (0, 255, 0)  # green
        thickness = 2
        img = cv2.rectangle(img, (x_min, y_min), (x_max, y_max), color, thickness)

    # Display the image with bounding boxes
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
    

# Example usage
draw_bbox("D:/FYP/all_training/real_positive_samples/Ash dieback Richard Buggs (1).jpg", "D:/FYP/all_training/yolo_label_real_positive/Ash dieback Richard Buggs (1).txt")
```
